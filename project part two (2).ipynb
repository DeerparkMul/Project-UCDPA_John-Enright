```python



GDP, score = result6['2018'], result6['2019 Score']
population, region = result6['Population (Millions)'], result6['Region']


plt.scatter(GDP, score, label=None,
            c=np.log10(population),
            s=region, linewidth=0, alpha=0.5)
plt.axis(aspect='equal')
plt.xlabel('2018 GDP')
plt.ylabel('2019 Score')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# Here we create a legend:
# we'll plot empty lists with the desired size and label
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area,
                label=str(region) + ' km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='Region')

plt.title('GDP VS Score: Population and Region');
```
